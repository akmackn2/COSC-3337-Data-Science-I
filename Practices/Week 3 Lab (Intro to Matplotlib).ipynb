```python
#Plotting Numpy Arrays
```

```python
x = np.linspace(0,15)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title('My Graph')
plt.xlabel('The x-axis')
plt.ylabel('The y-axis')
plt.show()
```

```python
plt.plot(x, np.sin(x), label= 'Y = Sin(x)')
plt.plot(x, np.cos(x), label= 'Y = Cos(x)')
plt.title('My Graph')
plt.xlabel('The x-axis')
plt.ylabel('The y-axis')
plt.legend()
plt.show()
```

```python
#Subplots
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, np.sin(x))
axes[0].set_title('Y=Sin(x)')
axes[1].plot(x, np.cos(x))
axes[1].set_title('Y=Cos(x)')

plt.show()
```

```python
x = np.linspace(1,25)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14,8))
axes[0][0].plot(x, x**3)
axes[0][1].plot(x, x)
axes[1][0].plot(x, np.log(x))
axes[1][1].plot(x, x**5)
plt.show()
```

```python
#Custom Axes Range
```

```python
x = np.linspace(0,5,11)
fig, axes = plt.subplots(nrows=1,ncols=3)

axes[0].plot(x,x**2,x**4)
axes[0].set_title('Default')

axes[1].plot(x,x**2,x**4)
axes[1].axis('tight')
axes[1].set_title('Tight')

axes[2].plot(x,x**2,x**4)
axes[2].set_xlim([1,5])
axes[2].set_ylim([0,40])
axes[2].set_title('Limited')

plt.show()
```

```python
pokemon_df['Type 1'].value_counts()
```

```python
type_1_names = pokemon_df['Type 1'].value_counts().index
type_1_values = pokemon_df['Type 1'].value_counts().values
plt.bar(type_1_names, type_1_values)
plt.title('Type 1 Pokemon Value')
plt.show()
```

```python
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral','brown']
patches, texts = plt.pie(type_1_values[:5], colors=colors, shadow=True, startangle=90)
plt.legend(patches, type_1_names[:5])
plt.axis('equal')
plt.tight_layout()
plt.show()
```
