```python
customer_df = pd.get_dummies(customer_df, drop_first=True)
sns.heatmap(customer_df.corr(), annot=True)
plt.show()
```

```python
plt.pie(customer_df['Genre_Male'].value_counts(), labels=['Female','Male'], colors=['Pink','Blue'], shadow=True, explode=[0, 0.1])
plt.legend()
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
x = customer_df.iloc[:, [2, 3]].values
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=3, min_samples=4, metric='euclidean')
```

```python
db.fit(x)
```

```python
db.labels_
```

```python
y_predictions = db.fit_predict(x)
```

```python
plt.scatter(x[y_predictions == 0, 0], x[y_predictions == 0, 1], s = 50, c = 'pink', label = 'cluster 1')
plt.scatter(x[y_predictions == 1, 0], x[y_predictions == 1, 1], s = 50, c = 'yellow', label = 'cluster 1')
plt.scatter(x[y_predictions == 2, 0], x[y_predictions == 2, 1], s = 50, c = 'cyan', label = 'cluster 1')
plt.scatter(x[y_predictions == 3, 0], x[y_predictions == 3, 1], s = 50, c = 'magenta', label = 'cluster 1')
plt.scatter(x[y_predictions == 4, 0], x[y_predictions == 4, 1], s = 50, c = 'orange', label = 'cluster 1')
plt.scatter(x[y_predictions == 5, 0], x[y_predictions == 5, 1], s = 50, c = 'blue', label = 'cluster 1')
plt.scatter(x[y_predictions == 6, 0], x[y_predictions == 6, 1], s = 50, c = 'red', label = 'cluster 1')
plt.scatter(x[y_predictions == 7, 0], x[y_predictions == 7, 1], s = 50, c = 'black', label = 'cluster 1')
plt.scatter(x[y_predictions == 8, 0], x[y_predictions == 8, 1], s = 50, c = 'violet', label = 'cluster 1')
plt.title('Clusters of customers')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```
