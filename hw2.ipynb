```python
#The two types of models I can use for my findings is a Logistic Regression & Random-Forest.
#Label encoder is used for converting categorical values into a unique number which is best for use in ordinal datasets.
#One-Hot-Encoding is used for dataset involving logistic regression in converting categorical value to a binary either 0 or 1.
#Multicollinearity is when multiple independent variables shows a strong relation/correlation to one another.
#Scaling data is important as it plays a role in improving performance of our model as well as reducing any bias that can degrade our model.
#We can handle missing values by either dropping them, replacing them with NaN, or making other adjustments that can help our model be better.
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import (confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report)
```

```python
x = data_df.drop(columns = 'output')
```

```python
y = data_df['output']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42, stratify = y)
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
cv = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
logReg_model = LogisticRegression(random_state = 42)
```

```python
scores = cross_val_score(logReg_model, x_train, y_train, cv=cv, scoring='accuracy', n_jobs=-1)
```

```python
print("Cross-Validation Accuracy: {:.3f} ± {:.3f}".format(np.mean(scores), np.std(scores)))
```

```python
logReg_model.fit(x_train, y_train)
```

```python
y_pred = logReg_model.predict(x_test)
```

```python
confusion_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(confusion_matrix, annot=True)
plt.title('Standard Binary Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
metrics = {
    "Accuracy Score": accuracy_score(y_test, y_pred),
    "Precision Score": precision_score(y_test, y_pred),
    "Recall Score": recall_score(y_test, y_pred),
    "F1 Score Score": f1_score(y_test, y_pred),
}
```

```python
for metric, value in metrics.items():
    print(f"{metric}: {value:.2f}")
```

```python
print(classification_report(y_test, y_pred))
```

```python
#Based on what's presented from Logistic Regression given the various scores, the scores indicate good results.
#Despite the good results, its important to also try another type of model for experimenting better results.
#It never hurts to try another model, especially when it can importantly play a role in some way for our type of dataset.
#It could be possible that trying Random-Forest Model could yield slightly better results making it a better model.
#Doing this can help make comparisons, learn what models to use for similar datasets in the future, etc.
```
